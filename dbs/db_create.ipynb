```python
cur.execute("""DROP TABLE IF EXISTS feed_sources;""")
conn.commit()
```

```python
cur.execute("""CREATE TABLE feed_sources AS
  SELECT t.*, '' sfunc
  FROM feed_sources_temp t;""")
conn.commit()
```

```python
cur.execute("""CREATE TABLE IF NOT EXISTS feed_sources(
   fsid INTEGER PRIMARY KEY AUTOINCREMENT,
   sname TEXT,
   slink TEXT,
   rss_sw INT DEFAULT 0,
   stitle TEXT,
   description TEXT,
   slink_img TEXT,
   sfunc TEXT,
   rdate TEXT);
""")
conn.commit()

cur.execute("""CREATE TABLE IF NOT EXISTS feed_entries(
                                feid INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT,
                                fsid INT,
                                u_etitle TEXT,
                                u_summary TEXT,
                                etitle TEXT,
                                summary TEXT,
                                elink TEXT,
                                published TEXT,
                                elink_img TEXT,
                                score INT,
                                group_topic TEXT,
                                topic TEXT,
                                group_theme TEXT,
                                theme TEXT,
                                project TEXT,
                                uname TEXT,
                                rdate TEXT);
                                """)
conn.commit()
```
